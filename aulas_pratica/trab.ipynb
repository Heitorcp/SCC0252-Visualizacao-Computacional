```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
imdb = pd.read_csv("../datasets/imdb/imdb_top_1000.csv")
```

## Relacao entre genero e preco

```python
imdb.Gross = imdb.Gross.str.replace(",","")
imdb.Gross = imdb.Gross.astype(float)
```

```python
imdb["gender01"] = imdb.Genre.str.split(",", expand=True).iloc[:,[0]]
imdb["gender02"] = imdb.Genre.str.split(",", expand=True).iloc[:,[1]]
imdb["gender03"] = imdb.Genre.str.split(",", expand=True).iloc[:,[2]]
```

```python
imdb
```

## Revenue per Director 

```python
directors = imdb[["Director", "Gross", "IMDB_Rating"]]
directors.sort_values(by="Director", ascending=True)
```

```python
directors_unique = directors.Director.unique()
len(directors_unique)
```

```python
directors
```

```python
director_count = directors.Director.value_counts().reset_index()
director_count_dict = dict(zip(director_count["index"], director_count.Director))
```

```python
director_count_dict
```

```python
directors["Count"] = directors.Director.map(director_count_dict)
```

```python
directors
```

## Mean rating per director


```python
directors_mean = directors.groupby("Director")["IMDB_Rating"].mean().reset_index()
directors_mean = directors_mean.sort_values(by=directors_mean.columns[1], ascending=False)
```
